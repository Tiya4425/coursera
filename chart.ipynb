```python
#pie chart
import numpy as np
import matplotlib.pyplot as plt
labels = ['Strawberry','Mango','Rasberry','Guava']
sizes = [10,20,50,30]

#แบ่งพิซซ่าออกมาให้ชัดๆ
explode = (0.1,0.1,0.1,0.3)

fig , ax=plt.subplots()
ax.pie(sizes ,
        explode = explode,
      labels=labels,
      
      autopct ='%1.1f%%',
      shadow = True,
      startangle = 90)
ax.axis('equal')
ax.set_title ('The number of Fruits')

#save รูป
plt.savefig('pie_chart.png')
plt.show()
```

```python
# bar chart
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
Fruits= ['Strawberry','Mango','Rasberry','Guava']
sizes = [10,20,50,30]

fig , ax=plt.subplots()
ax.bar(Fruits ,sizes,align='center' , color=['black', 'red', 'green', 'blue'],edgecolor='blue')
ax.set_ylabel('Fruits')
ax.set_ylabel('the number of fruits')
```

```python
import matplotlib.pyplot as plt
Year = ['2001' ,'2002','2003' ,'2004']
Sales = [15.2 ,15.1 ,14.3 ,14.2]
plt.plot (Year,Sales ,color='red',marker='o')
plt.title('Sales from 2001-2004')
plt.xlabel('Year',fontsize =12)
plt.ylabel('Sales',fontsize=12)
plt.grid(True)
plt.show()
```
